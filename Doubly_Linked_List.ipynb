```python
class train:
  def __init__(self,prev=None,coach=None,next=None):
    self.prev = prev
    self.coach = coach
    self.next = next

class DLL:
  def __init__(self,head=None):
    self.head = head

  def is_empty(self):
    return self.head==None

  def insert_beg(self,coach):
    new_coach = train(None,coach,self.head)
    if not self.is_empty():
      temp = self.head
      temp.prev = new_coach
    self.head = new_coach

  def insert_end(self,coach):
    if not self.is_empty():
      temp = self.head
      while temp.next is not None:
        temp = temp.next
      temp.next = train(temp,coach,None)
      return
    else:
      self.head = train(None,coach,None)
      return

  def print_list(self):
    temp = self.head
    while temp is not None:
      print(temp.coach,end=" ")
      temp = temp.next

  def delete_beg(self):
    if self.head is not None:
      temp = self.head
      self.head = temp.next
      if self.head is not None:
        self.head.prev = None
        return
    return

  def delete_end(self):
    if not self.is_empty():
      temp = self.head
      while temp.next.next is not None:
        temp = temp.next
      temp.next = None

  def search(self,coach):
    if not self.is_empty():
      temp = self.head
    while temp is not None:
      if(temp.coach==coach):
        print("Element is Found: ",coach)
        return
      else:
        temp = temp.next

  def insert_after(self,coach):
    if not self.is_empty():
      temp = self.head
      if temp.next is None:
        if(temp.coach==coach):
          n = int(input())
          new = train(temp,n,None)
          temp.next = new
          return
      else:
        while temp.next is not None:
          if(temp.coach == coach):
            n = int(input())
            new = train(temp,n,temp.next)
            temp.next = new
            return
          else:
            temp = temp.next
    return None

  def delete_after(self,coach):
    if not self.is_empty():
      temp = self.head
      if temp.next is None:
          return None
      else:
        while temp.next is not None:
          if(temp.coach==coach):
            temp.next = temp.next.next
            return
          else:
            temp=temp.next
    return None


dll = DLL()
print(dll.is_empty())
dll.insert_beg(10)
dll.insert_beg(20)
dll.insert_end(30)
dll.print_list()
print("\n")
dll.delete_beg()
print("Delete_beg")
dll.print_list()
print("\n")
print("Delete at end")
dll.delete_end()
dll.print_list()
print("\n")
dll.search(10)
dll.insert_end(50)
dll.insert_end(70)
print("Inserting between")
dll.insert_after(10)
print("\n")
dll.print_list()
print("\n")
print("Deleting after this element: ")
dll.delete_after(10)
dll.print_list()
```
