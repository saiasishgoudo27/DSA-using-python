```python
class book:
  def __init__(self,title=None,next=None):
    self.title = title
    self.next = next

class SLL:
  def __init__(self,head=None):
    self.head = head

  def insert_beg(self,title):
    new_book = book(title,self.head)
    self.head = new_book

  def is_empty(self):
    return self.head==None

  def insert_end(self,title):
    new_book = book(title)
    if not self.is_empty():
        temp = self.head
        while temp.next is not None:
          temp = temp.next
        temp.next = new_book
    else:
      self.head = new_book

  def search(self,title):
    if not self.is_empty():
      temp = self.head
      while temp is not None:
        if(temp.title!=title):
          temp = temp.next
        else:
          print("Element Found: ",title)
          return
      print("Element is not Found")

  def print_SLL(self):
    temp = self.head
    while temp is not None:
      print(temp.title,end = " ")
      temp = temp.next
    return None

  def insert_after(self,title):
    if not self.is_empty():
      temp = self.head
      while temp is not None:
        if(temp.title==title):
          n = int(input())
          new_node = book(n,temp.next)
          temp.next = new_node
          return
        else:
          temp = temp.next
      print("Element is not Found")
    else:
      return None

  def delete_begin(self):
    if not self.is_empty():
      temp = self.head
      self.head = temp.next
      return
    else:
      return None
  def delete_end(self):
    if not self.is_empty():
      temp = self.head
      if temp.next is None:
        self.head=None
        return
      else:
        while temp.next.next is not None:
          temp = temp.next
        temp.next=None
        return
  def delete_after(self,title):
    if not self.is_empty():
      temp = self.head
      while temp.next is not None:
        if(title == temp.title):
          temp.next=temp.next.next
          return
        else:
          temp = temp.next
      return None



sll = SLL()
print(sll.is_empty())
sll.insert_beg(1)
sll.insert_beg(2)
sll.insert_beg(3)
sll.insert_end(5)
sll.insert_end(7)
sll.print_SLL()
print(sll.is_empty())
sll.search(1)
sll.insert_after(1)
sll.print_SLL()
print("\n")
sll.delete_begin()
sll.print_SLL()
sll.delete_end()
print("\n")
sll.print_SLL()
sll.delete_after(2)
print("\n")
sll.print_SLL()
```
